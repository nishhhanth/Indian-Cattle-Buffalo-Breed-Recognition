```python
import torch
from torchvision import datasets, transforms
from torch.utils.data import DataLoader

# Paths
TRAIN_DIR = "/kaggle/working/data_clean/train"
VAL_DIR   = "/kaggle/working/data_clean/val"

# Data augmentation (TRAIN)
train_transforms = transforms.Compose([
    transforms.RandomResizedCrop(224),
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(10),
    transforms.ColorJitter(brightness=0.2, contrast=0.2),
    transforms.ToTensor(),
    transforms.Normalize(
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225]
    )
])

# Validation transforms (NO augmentation)
val_transforms = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225]
    )
])

# Datasets
train_dataset = datasets.ImageFolder(TRAIN_DIR, transform=train_transforms)
val_dataset   = datasets.ImageFolder(VAL_DIR, transform=val_transforms)

# DataLoaders (GPU-safe batch size)
train_loader = DataLoader(
    train_dataset,
    batch_size=32,
    shuffle=True,
    num_workers=2,
    pin_memory=True
)

val_loader = DataLoader(
    val_dataset,
    batch_size=32,
    shuffle=False,
    num_workers=2,
    pin_memory=True
)

# Quick sanity check
print("✅ Train batches:", len(train_loader))
print("✅ Validation batches:", len(val_loader))
print("✅ Classes:", train_dataset.classes)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import models

# Device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Load EfficientNet-B3 pretrained
model = models.efficientnet_b3(pretrained=True)

# Replace classifier head
num_features = model.classifier[1].in_features
model.classifier[1] = nn.Linear(num_features, 15)

model = model.to(device)

# Loss
criterion = nn.CrossEntropyLoss()

# Optimizer
optimizer = optim.AdamW(model.parameters(), lr=1e-4)

# ✅ FIXED scheduler (NO verbose argument)
scheduler = optim.lr_scheduler.ReduceLROnPlateau(
    optimizer,
    mode="max",
    patience=3,
    factor=0.5
)

print("✅ Step 6 fixed: Model, optimizer, scheduler initialized")
```

```python
import torch
from torch.cuda.amp import autocast, GradScaler

scaler = GradScaler()

epochs = 30
best_val_acc = 0.0
patience = 5
early_stop_counter = 0

for epoch in range(epochs):

    # ---------------- TRAIN ----------------
    model.train()
    running_loss = 0.0

    for images, labels in train_loader:
        images = images.to(device, non_blocking=True)
        labels = labels.to(device, non_blocking=True)

        optimizer.zero_grad()

        with autocast():
            outputs = model(images)
            loss = criterion(outputs, labels)

        scaler.scale(loss).backward()
        scaler.step(optimizer)
        scaler.update()

        running_loss += loss.item()

    train_loss = running_loss / len(train_loader)

    # ---------------- VALIDATION ----------------
    model.eval()
    correct = 0
    total = 0

    with torch.no_grad():
        for images, labels in val_loader:
            images = images.to(device, non_blocking=True)
            labels = labels.to(device, non_blocking=True)

            with autocast():
                outputs = model(images)

            preds = torch.argmax(outputs, dim=1)
            correct += (preds == labels).sum().item()
            total += labels.size(0)

    val_acc = correct / total

    # Scheduler step (SAFE)
    scheduler.step(val_acc)

    print(
        f"Epoch [{epoch+1}/{epochs}] | "
        f"Train Loss: {train_loss:.4f} | "
        f"Val Acc: {val_acc:.4f}"
    )

    # ---------------- EARLY STOPPING ----------------
    if val_acc > best_val_acc:
        best_val_acc = val_acc
        torch.save(model.state_dict(), "best_efficientnet_b3.pt")
        early_stop_counter = 0
        print("✅ Best model saved")
    else:
        early_stop_counter += 1
        print(f"⏳ Early stopping counter: {early_stop_counter}/{patience}")

        if early_stop_counter >= patience:
            print("🛑 Early stopping triggered")
            break
```

```python
import torch
import torch.nn as nn
from torchvision import models

# Device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# ---------------- CNN Backbone ----------------
cnn = models.efficientnet_b3(pretrained=True)
cnn.classifier = nn.Identity()  # remove classifier head

# Output feature size = 1536
cnn_out_features = 1536

# ---------------- ViT Backbone ----------------
vit = models.vit_b_16(pretrained=True)
vit.heads = nn.Identity()  # remove classification head

# Output feature size = 768
vit_out_features = 768

# ---------------- Hybrid Network ----------------
class HybridCNNViT(nn.Module):
    def __init__(self, cnn, vit, num_classes):
        super().__init__()
        self.cnn = cnn
        self.vit = vit

        self.fc = nn.Sequential(
            nn.Linear(cnn_out_features + vit_out_features, 512),
            nn.BatchNorm1d(512),
            nn.ReLU(),
            nn.Dropout(0.5),
            nn.Linear(512, num_classes)
        )

    def forward(self, x):
        cnn_feat = self.cnn(x)
        vit_feat = self.vit(x)
        combined = torch.cat((cnn_feat, vit_feat), dim=1)
        return self.fc(combined)

# Create model
hybrid_model = HybridCNNViT(cnn, vit, num_classes=15).to(device)

print("✅ Hybrid CNN + ViT model initialized")
```

```python
import torch.optim as optim
import torch.nn as nn

# -------- Freeze ViT backbone initially --------
for param in hybrid_model.vit.parameters():
    param.requires_grad = False

# -------- Loss --------
criterion_hybrid = nn.CrossEntropyLoss()

# -------- Optimizer (train CNN + fusion head only) --------
optimizer_hybrid = optim.AdamW(
    filter(lambda p: p.requires_grad, hybrid_model.parameters()),
    lr=5e-5,        # lower LR for hybrid
    weight_decay=1e-4
)

# -------- Scheduler (NO verbose arg) --------
scheduler_hybrid = optim.lr_scheduler.ReduceLROnPlateau(
    optimizer_hybrid,
    mode="max",
    patience=3,
    factor=0.5
)

print("✅ Hybrid optimizer & scheduler initialized")
print("🔒 ViT backbone frozen")
```

```python
# ============================================================================
# COMPLETE MODEL EVALUATION SCRIPT FOR CATTLE BREED CLASSIFICATION
# Paste this code into new cells in your Jupyter notebook
# ============================================================================

# Cell 1: Install required packages (if needed)
# !pip install seaborn scikit-learn matplotlib pandas

# Cell 2: Import all required libraries
import torch
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import (
    confusion_matrix, classification_report,
    roc_curve, auc, precision_recall_curve,
    average_precision_score
)
from sklearn.preprocessing import label_binarize
import pandas as pd
from torch.utils.data import DataLoader
import os

# Set plotting style
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")

# Create output directory
os.makedirs('/kaggle/working/evaluation_plots', exist_ok=True)

# Cell 3: Helper functions for all visualizations

def plot_confusion_matrix_svg(y_true, y_pred, class_names, model_name, save_path):
    """Generate confusion matrix heatmap"""
    cm = confusion_matrix(y_true, y_pred)
    cm_normalized = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]

    fig, axes = plt.subplots(1, 2, figsize=(24, 10))

    # Raw counts
    sns.heatmap(cm, annot=True, fmt='d', cmap='YlOrRd',
                xticklabels=class_names, yticklabels=class_names,
                cbar_kws={'label': 'Count'}, ax=axes[0], linewidths=0.5)
    axes[0].set_title(f'{model_name} - Confusion Matrix (Counts)',
                      fontsize=16, fontweight='bold', pad=20)
    axes[0].set_ylabel('True Label', fontsize=13, fontweight='bold')
    axes[0].set_xlabel('Predicted Label', fontsize=13, fontweight='bold')
    axes[0].tick_params(axis='both', labelsize=9)
    plt.setp(axes[0].get_xticklabels(), rotation=45, ha='right')
    plt.setp(axes[0].get_yticklabels(), rotation=0)

    # Normalized
    sns.heatmap(cm_normalized, annot=True, fmt='.2f', cmap='Blues',
                xticklabels=class_names, yticklabels=class_names,
                cbar_kws={'label': 'Proportion'}, ax=axes[1],
                vmin=0, vmax=1, linewidths=0.5)
    axes[1].set_title(f'{model_name} - Confusion Matrix (Normalized)',
                      fontsize=16, fontweight='bold', pad=20)
    axes[1].set_ylabel('True Label', fontsize=13, fontweight='bold')
    axes[1].set_xlabel('Predicted Label', fontsize=13, fontweight='bold')
    axes[1].tick_params(axis='both', labelsize=9)
    plt.setp(axes[1].get_xticklabels(), rotation=45, ha='right')
    plt.setp(axes[1].get_yticklabels(), rotation=0)

    plt.tight_layout()
    plt.savefig(save_path, format='svg', dpi=300, bbox_inches='tight')
    plt.close()
    print(f"✅ Confusion matrix saved: {save_path}")

def plot_roc_curves_svg(y_true, y_pred_proba, class_names, model_name, save_path):
    """Generate ROC curves for all classes"""
    n_classes = len(class_names)
    y_true_bin = label_binarize(y_true, classes=range(n_classes))

    fpr, tpr, roc_auc = {}, {}, {}

    fig, ax = plt.subplots(figsize=(14, 12))
    colors = plt.cm.tab20(np.linspace(0, 1, n_classes))

    for i in range(n_classes):
        fpr[i], tpr[i], _ = roc_curve(y_true_bin[:, i], y_pred_proba[:, i])
        roc_auc[i] = auc(fpr[i], tpr[i])
        ax.plot(fpr[i], tpr[i], color=colors[i], linewidth=2.5,
                label=f'{class_names[i]} (AUC = {roc_auc[i]:.3f})')

    ax.plot([0, 1], [0, 1], 'k--', linewidth=2.5, label='Random (AUC = 0.500)')
    ax.set_xlim([0.0, 1.0])
    ax.set_ylim([0.0, 1.05])
    ax.set_xlabel('False Positive Rate', fontsize=14, fontweight='bold')
    ax.set_ylabel('True Positive Rate', fontsize=14, fontweight='bold')
    ax.set_title(f'{model_name} - ROC Curves (One-vs-Rest)',
                 fontsize=16, fontweight='bold', pad=20)
    ax.legend(loc='lower right', fontsize=9, ncol=2, framealpha=0.9)
    ax.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.savefig(save_path, format='svg', dpi=300, bbox_inches='tight')
    plt.close()
    print(f"✅ ROC curves saved: {save_path}")
    return roc_auc

def plot_pr_curves_svg(y_true, y_pred_proba, class_names, model_name, save_path):
    """Generate Precision-Recall curves"""
    n_classes = len(class_names)
    y_true_bin = label_binarize(y_true, classes=range(n_classes))

    precision, recall, avg_precision = {}, {}, {}

    fig, ax = plt.subplots(figsize=(14, 12))
    colors = plt.cm.tab20(np.linspace(0, 1, n_classes))

    for i in range(n_classes):
        precision[i], recall[i], _ = precision_recall_curve(
            y_true_bin[:, i], y_pred_proba[:, i]
        )
        avg_precision[i] = average_precision_score(
            y_true_bin[:, i], y_pred_proba[:, i]
        )
        ax.plot(recall[i], precision[i], color=colors[i], linewidth=2.5,
                label=f'{class_names[i]} (AP = {avg_precision[i]:.3f})')

    ax.set_xlim([0.0, 1.0])
    ax.set_ylim([0.0, 1.05])
    ax.set_xlabel('Recall', fontsize=14, fontweight='bold')
    ax.set_ylabel('Precision', fontsize=14, fontweight='bold')
    ax.set_title(f'{model_name} - Precision-Recall Curves',
                 fontsize=16, fontweight='bold', pad=20)
    ax.legend(loc='best', fontsize=9, ncol=2, framealpha=0.9)
    ax.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.savefig(save_path, format='svg', dpi=300, bbox_inches='tight')
    plt.close()
    print(f"✅ PR curves saved: {save_path}")
    return avg_precision

def create_error_analysis_svg(y_true, y_pred, class_names, model_name, save_path):
    """Generate error analysis table"""
    cm = confusion_matrix(y_true, y_pred)

    error_data = []
    for i, class_name in enumerate(class_names):
        total = cm[i].sum()
        correct = cm[i, i]
        errors = total - correct
        error_rate = errors / total if total > 0 else 0
        accuracy = correct / total if total > 0 else 0

        confused_with = []
        for j in range(len(class_names)):
            if i != j and cm[i, j] > 0:
                confused_with.append((class_names[j], cm[i, j]))

        confused_with.sort(key=lambda x: x[1], reverse=True)
        top_confusion = confused_with[0] if confused_with else ("None", 0)

        error_data.append({
            'Class': class_name,
            'Total': total,
            'Correct': correct,
            'Errors': errors,
            'Accuracy': f'{accuracy:.1%}',
            'Error Rate': f'{error_rate:.1%}',
            'Top Confusion': f'{top_confusion[0]} ({top_confusion[1]})'
        })

    df = pd.DataFrame(error_data)

    fig, ax = plt.subplots(figsize=(18, 12))
    ax.axis('tight')
    ax.axis('off')

    table = ax.table(cellText=df.values,
                     colLabels=df.columns,
                     cellLoc='center',
                     loc='center',
                     bbox=[0, 0, 1, 1])

    table.auto_set_font_size(False)
    table.set_fontsize(10)
    table.scale(1.2, 2.5)

    for i in range(len(df.columns)):
        table[(0, i)].set_facecolor('#2c3e50')
        table[(0, i)].set_text_props(weight='bold', color='white')

    for i in range(1, len(df) + 1):
        if i % 2 == 0:
            for j in range(len(df.columns)):
                table[(i, j)].set_facecolor('#ecf0f1')

    plt.title(f'{model_name} - Error Analysis by Class',
              fontsize=18, fontweight='bold', pad=30)
    plt.savefig(save_path, format='svg', dpi=300, bbox_inches='tight')
    plt.close()
    print(f"✅ Error analysis saved: {save_path}")

# Cell 4: Main evaluation function

def comprehensive_model_evaluation(model, test_loader, class_names,
                                   model_name, device):
    """Run complete evaluation and generate all plots"""

    model.eval()
    all_preds, all_labels, all_probs = [], [], []

    print(f"\n{'='*60}")
    print(f"🔍 EVALUATING {model_name}")
    print(f"{'='*60}\n")

    with torch.no_grad():
        for images, labels in test_loader:
            images = images.to(device)
            outputs = model(images)
            probs = torch.softmax(outputs, dim=1)
            preds = torch.argmax(outputs, dim=1)

            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.numpy())
            all_probs.extend(probs.cpu().numpy())

    all_probs = np.array(all_probs)

    # Generate visualizations
    base_path = f'/kaggle/working/evaluation_plots/{model_name}'

    print("📊 Generating confusion matrix...")
    plot_confusion_matrix_svg(all_labels, all_preds, class_names, model_name,
                              f'{base_path}_confusion_matrix.svg')

    print("📈 Generating ROC curves...")
    roc_auc = plot_roc_curves_svg(all_labels, all_probs, class_names, model_name,
                                   f'{base_path}_roc_curves.svg')

    print("📉 Generating PR curves...")
    avg_precision = plot_pr_curves_svg(all_labels, all_probs, class_names, model_name,
                                       f'{base_path}_pr_curves.svg')

    print("📋 Generating error analysis...")
    create_error_analysis_svg(all_labels, all_preds, class_names, model_name,
                              f'{base_path}_error_analysis.svg')

    # Get classification report
    report = classification_report(all_labels, all_preds,
                                  target_names=class_names,
                                  output_dict=True)

    # Print summary
    print(f"\n{'='*60}")
    print(f"📊 {model_name} SUMMARY METRICS")
    print(f"{'='*60}")
    print(f"Overall Accuracy:    {report['accuracy']:.4f}")
    print(f"Macro Avg F1-Score:  {report['macro avg']['f1-score']:.4f}")
    print(f"Weighted Avg F1:     {report['weighted avg']['f1-score']:.4f}")
    print(f"Mean ROC-AUC:        {np.mean(list(roc_auc.values())):.4f}")
    print(f"Mean AP:             {np.mean(list(avg_precision.values())):.4f}")
    print(f"{'='*60}\n")

    return report, roc_auc, avg_precision

# Cell 5: Model comparison functions

def plot_model_comparison(report1, report2, class_names, save_path):
    """Compare metrics between two models"""
    metrics = ['precision', 'recall', 'f1-score']

    fig, axes = plt.subplots(1, 3, figsize=(24, 7))

    for idx, metric in enumerate(metrics):
        model1_scores = [report1[cls][metric] for cls in class_names]
        model2_scores = [report2[cls][metric] for cls in class_names]

        x = np.arange(len(class_names))
        width = 0.35

        bars1 = axes[idx].bar(x - width/2, model1_scores, width,
                             label='EfficientNet-B3', color='#e74c3c',
                             alpha=0.8, edgecolor='black', linewidth=1.2)
        bars2 = axes[idx].bar(x + width/2, model2_scores, width,
                             label='Hybrid CNN+ViT', color='#2ecc71',
                             alpha=0.8, edgecolor='black', linewidth=1.2)

        axes[idx].set_xlabel('Breed Class', fontsize=13, fontweight='bold')
        axes[idx].set_ylabel(metric.capitalize(), fontsize=13, fontweight='bold')
        axes[idx].set_title(f'{metric.capitalize()} Comparison by Class',
                           fontsize=15, fontweight='bold', pad=15)
        axes[idx].set_xticks(x)
        axes[idx].set_xticklabels(class_names, rotation=45, ha='right', fontsize=9)
        axes[idx].legend(fontsize=11, loc='lower right')
        axes[idx].grid(True, alpha=0.3, axis='y', linestyle='--')
        axes[idx].set_ylim([0, 1.05])
        axes[idx].axhline(y=0.5, color='gray', linestyle='--', alpha=0.5)

    plt.tight_layout()
    plt.savefig(save_path, format='svg', dpi=300, bbox_inches='tight')
    plt.close()
    print(f"✅ Model comparison saved: {save_path}")

def create_performance_comparison_table(report1, report2, roc1, roc2, ap1, ap2, save_path):
    """Create comprehensive comparison table"""

    data = {
        'Metric': [
            'Overall Accuracy',
            'Macro Precision',
            'Macro Recall',
            'Macro F1-Score',
            'Weighted Precision',
            'Weighted Recall',
            'Weighted F1-Score',
            'Mean ROC-AUC',
            'Mean Average Precision'
        ],
        'EfficientNet-B3': [
            f"{report1['accuracy']:.4f}",
            f"{report1['macro avg']['precision']:.4f}",
            f"{report1['macro avg']['recall']:.4f}",
            f"{report1['macro avg']['f1-score']:.4f}",
            f"{report1['weighted avg']['precision']:.4f}",
            f"{report1['weighted avg']['recall']:.4f}",
            f"{report1['weighted avg']['f1-score']:.4f}",
            f"{np.mean(list(roc1.values())):.4f}",
            f"{np.mean(list(ap1.values())):.4f}"
        ],
        'Hybrid CNN+ViT': [
            f"{report2['accuracy']:.4f}",
            f"{report2['macro avg']['precision']:.4f}",
            f"{report2['macro avg']['recall']:.4f}",
            f"{report2['macro avg']['f1-score']:.4f}",
            f"{report2['weighted avg']['precision']:.4f}",
            f"{report2['weighted avg']['recall']:.4f}",
            f"{report2['weighted avg']['f1-score']:.4f}",
            f"{np.mean(list(roc2.values())):.4f}",
            f"{np.mean(list(ap2.values())):.4f}"
        ]
    }

    # Calculate improvements
    improvements = []
    for i in range(len(data['Metric'])):
        val1 = float(data['EfficientNet-B3'][i])
        val2 = float(data['Hybrid CNN+ViT'][i])
        diff = val2 - val1
        improvements.append(f"{diff:+.4f}")

    data['Improvement'] = improvements

    df = pd.DataFrame(data)

    fig, ax = plt.subplots(figsize=(16, 10))
    ax.axis('tight')
    ax.axis('off')

    table = ax.table(cellText=df.values,
                     colLabels=df.columns,
                     cellLoc='center',
                     loc='center',
                     bbox=[0, 0, 1, 1])

    table.auto_set_font_size(False)
    table.set_fontsize(11)
    table.scale(1.3, 3)

    # Header styling
    for i in range(len(df.columns)):
        cell = table[(0, i)]
        cell.set_facecolor('#2c3e50')
        cell.set_text_props(weight='bold', color='white', fontsize=12)

    # Row styling
    for i in range(1, len(df) + 1):
        # Metric name column
        table[(i, 0)].set_facecolor('#95a5a6')
        table[(i, 0)].set_text_props(weight='bold')

        # Color improvement column
        improvement = float(df.iloc[i-1]['Improvement'])
        if improvement > 0:
            table[(i, 3)].set_facecolor('#d5f4e6')
            table[(i, 3)].set_text_props(color='#27ae60', weight='bold')
        elif improvement < 0:
            table[(i, 3)].set_facecolor('#fadbd8')
            table[(i, 3)].set_text_props(color='#c0392b', weight='bold')

    plt.title('Comprehensive Model Performance Comparison',
              fontsize=18, fontweight='bold', pad=30)
    plt.savefig(save_path, format='svg', dpi=300, bbox_inches='tight')
    plt.close()
    print(f"✅ Performance comparison table saved: {save_path}")

# Cell 6: RUN EVALUATION FOR BOTH MODELS

print("\n" + "="*70)
print("🚀 STARTING COMPREHENSIVE MODEL EVALUATION")
print("="*70 + "\n")

# Evaluate EfficientNet-B3
efficientnet_report, eff_roc, eff_ap = comprehensive_model_evaluation(
    model,
    test_loader,
    train_dataset.classes,
    'EfficientNet_B3',
    device
)

# Evaluate Hybrid Model
hybrid_report, hyb_roc, hyb_ap = comprehensive_model_evaluation(
    hybrid_model,
    test_loader,
    train_dataset.classes,
    'Hybrid_CNN_ViT',
    device
)

# Generate comparison plots
print("\n" + "="*70)
print("📊 GENERATING MODEL COMPARISON VISUALIZATIONS")
print("="*70 + "\n")

plot_model_comparison(
    efficientnet_report,
    hybrid_report,
    train_dataset.classes,
    '/kaggle/working/evaluation_plots/models_comparison_metrics.svg'
)

create_performance_comparison_table(
    efficientnet_report,
    hybrid_report,
    eff_roc,
    hyb_roc,
    eff_ap,
    hyb_ap,
    '/kaggle/working/evaluation_plots/performance_comparison_table.svg'
)

print("\n" + "="*70)
print("✅ ALL EVALUATIONS COMPLETED SUCCESSFULLY!")
print("="*70)
print(f"\n📁 All plots saved in: /kaggle/working/evaluation_plots/")
print("\nGenerated files:")
print("  • EfficientNet_B3_confusion_matrix.svg")
print("  • EfficientNet_B3_roc_curves.svg")
print("  • EfficientNet_B3_pr_curves.svg")
print("  • EfficientNet_B3_error_analysis.svg")
print("  • Hybrid_CNN_ViT_confusion_matrix.svg")
print("  • Hybrid_CNN_ViT_roc_curves.svg")
print("  • Hybrid_CNN_ViT_pr_curves.svg")
print("  • Hybrid_CNN_ViT_error_analysis.svg")
print("  • models_comparison_metrics.svg")
print("  • performance_comparison_table.svg")
print("\n" + "="*70 + "\n")
```
